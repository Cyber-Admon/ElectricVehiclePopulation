```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv
```

Reading CSV File

```python
data = automobiles = read_csv("Electric_Vehicle_Population_Data.csv")
data.rename(columns={"Model Year": "Year"}, inplace=True)
```

Grouping Data by Model Year

```python
grouped_by_year = data.groupby('Year').size()
grouped_by_year.to_csv("GroupedByYear.csv", index=True)
```

Reading from the Grouped Data

```python
carByYear = read_csv("GroupedByYear.csv")
carByYear.rename(columns={"0": "Size"}, inplace=True)
print(carByYear.head())
```

Ploting a Bar Chart using Matplotlib 

```python
years = carByYear["Year"]
sizes = carByYear["Size"]

plt.bar(years, sizes)

plt.xlabel("Year")
plt.ylabel("Size")
plt.title("Electric Vehicle Population by Year")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()

plt.savefig("EVPG(bar).png", transparent=True) 
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
plt.plot(years, sizes, marker="o", linestyle="--", color="blue")


plt.xlabel("Year", labelpad=10)
plt.ylabel("Size", labelpad=10)
plt.title("Electric Vehicle Population Growth (1997-2002)", fontsize=16)


plt.grid(True, alpha=0.5)
plt.xticks(years)
plt.yticks(range(0, 10, 2))
plt.xlim(min(years), max(years))

plt.savefig("EVPG(line).png", transparent=True) 
# Show the plot
plt.show()
```

Population By County

```python
data = automobiles = read_csv("Electric_Vehicle_Population_Data.csv")
data.rename(columns={"Model Year": "Year"}, inplace=True)
print(data.head())
```

Grouping By County

```python
groupByCounty = data.groupby("County").size()
groupByCounty.to_csv("GroupByCounty.csv", index=True)
```

Read from the CSV File

```python
dataByCounty = read_csv("GroupByCounty.csv")
dataByCounty.rename(columns={"0": "Population"}, inplace=True)
print(dataByCounty)
```

```python
county = dataByCounty["County"]
population = dataByCounty["Population"]

plt.bar(county, population)

plt.xlabel("county")
plt.ylabel("population")
plt.title("Electric Vehicle Population by County")
plt.xticks(rotation=45, ha="right")

plt.savefig("EVPC(bar).png", transparent=True) 
```
